### Data Wrangling Code

````python
# Making sure that price = 0 is taken care of
temp = df.groupby('room_type')['price'].mean()
````

````python
temp
````

````python
df.loc[df['room_type'] == 'Entire home/apt',['price']]

# 25409 listings are of type entire home/apt
````

````python
# Replacing the price of listing having price = 0 with roomtype mean price

df.loc[(df['room_type'] == 'Entire home/apt') & (df['price'] == 0), 'price'] = temp[0]
df.loc[(df['room_type'] == 'Private room') & (df['price'] == 0), 'price'] = temp[1]
df.loc[(df['room_type'] == 'Shared room') & (df['price'] == 0), 'price'] = temp[2]
````

````python
# Since minimum_nights are the number of nights you get to live in a lodging for the price in the price column, we can create a df where we will be arranging the listings of every particular
#  roomtype according to the price of lodging for one day .We will be arranging the listing with lowest price per day to highest price per day for each roomtype category.For that we
# will create a new column 'price per day'.

df['price per day'] = df['price'] / df['minimum_nights']
df['price per day'] = df['price per day'].round(2)
df.head()

````
