### Data Wrangling Code

```python
# Making sure that price = 0 is taken care of
temp = df.groupby('room_type')['price'].mean()
```

```python
temp
```

```python
# The price range of the entire home/apt 

df.loc[df['room_type'] == 'Entire home/apt',['price']].sort_values(by = 'price' , ascending = False)

# 25409 listings are of type entire home/apt
```

```python
# Checking the listings with price = 0
print(len(df.loc[df['price'] == 0 , :]))
df.loc[df['price'] == 0 , :]
```

```python
# Replacing the price of listing having price = 0 with roomtype mean price

df.loc[(df['room_type'] == 'Entire home/apt') & (df['price'] == 0), 'price'] = temp[0]
df.loc[(df['room_type'] == 'Private room') & (df['price'] == 0), 'price'] = temp[1]
df.loc[(df['room_type'] == 'Shared room') & (df['price'] == 0), 'price'] = temp[2]
```
